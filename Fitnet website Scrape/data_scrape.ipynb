```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
ua = 'Mozilla/5.0 (Windows; U; Windows NT 5.1; en-US; rv:1.9.0.7) Gecko/2009021910 Firefox/3.0.7'
response = requests.get('https://www.fitnet.in/', headers={'User-Agent':'ua'})
response.status_code
```

```python
html = response.content
soup = BeautifulSoup(html, 'html.parser')
soup
```

```python
outer_div = soup.find('div', class_='collapse navbar-collapse')
outer_div
```

```python
centers_tags  = outer_div.find('ul', class_= 'dropdown-menu')
centers_tags
```

```python
lists  = centers_tags.find_all('li')
lists
```

```python
links = [listed.contents[0]['href'] for listed in lists]
links
```

```python
center_names  = [i.text for i in lists]
center_names
```

```python
link_to_name = dict(zip(links,center_names))
link_to_name
```

```python
link_to_name_df = pd.DataFrame({
    'Links':links,
    'Center Names':center_names
})
```

```python
link_to_name_df
```

### Center 1

```python
link1 = "https://fitnet.in/centers/centers-view/1"
response1 = requests.get(link1, headers={'User-Agent':'ua'})
response1.status_code
```

```python
html1 = response1.content
soup1 = BeautifulSoup(html1, 'lxml')
all_divs1 = soup1.find_all('div', class_='col-lg-3')
all_divs1
```

```python
location1 = soup1.find('p', class_='head-caption').text
location1
```

```python
all_divs1[0].find('div', class_='price-box-head')
```

```python
headings = [head.find('div', class_='price-box-head') for head in all_divs1]
headings1 = [head.find('h2').text for head in headings]
headings1
```

```python
prices = [head.find('div', class_='price-box-ctn') for head in all_divs1]
prices1 = [head.find('p').text.strip('₹ ') for head in prices]
prices1
```

```python
facilities = [head.find('ul', class_='price-list') for head in all_divs1]
facilities
```

```python
facility = []
for l in facilities:
    lists = l.find_all('li')
    content = [li.text for li in lists]
    facility.append(",  ".join(content))
facility
```

```python
first = pd.DataFrame({
    'Duration': headings1,
    'Charges': prices1,
    'Facilities':facility,
    'Location':location1
})
```

```python
pd.set_option('display.max_colwidth', None)
```

```python
first
```

```python

first[['Facility1', 'Facility2', 'Facility3', 'Facility4', 'Facility5']] = first["Facilities"].str.split(',',n=4, expand = True)
```

```python
first.drop(['Facilities'],axis=1)
```

```python
first.to_csv('First_location.csv', index=False)
```
